```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import csv
import pathlib as path
import plotly.express as px
```

```python
df=pd.read_csv("homicide_by_countries.csv", encoding='unicode_escape')
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df['Location']=df['Location'].astype(str)
```

```python
df['Rate']=df['Rate'].astype(int)
```

```python
my_list1=['Location','Region','Subregion']
for i in my_list1:
  print(i)
  df[i]=df[i].astype(str)
```

```python
my_list=['Rate','Count','Year']
for i in my_list:
  print(i)
  df[i]=df[i].astype(int)
```

```python
df.dtypes
```

```python
df['Region'].replace('Americas','N/S America',inplace=True)
```

```python
df
```

```python
df1=df.sort_values('Count',ascending=False)
df1
```

```python
df1=df[['Location','Count']].sort_values(by='Count',ascending=False).head(5)
df1['Perc']=(df['Count']*100/df1['Count'].sum()).round(2)
df1
```

```python
df1.plot(x='Location',y='Count',kind='pie',labels=df1.Location,autopct='%1.2f%%')
plt.legend().set_visible(False)
```

```python
df1.Location
```

```python
if not isinstance(df1, pd.DataFrame):
    raise TypeError("'df1' must be a pandas.DataFrame object.")
```

```python
df5=df.groupby(['Year'])['Rate'].sum().sort_values(ascending=False)
df5
```

```python
df5.plot(kind='bar',figsize=(7,3),color='skyblue',edgecolor='black')
plt.xlabel('Region Year')
plt.ylabel('Sum_of_Rate')
plt.title('Sum of rate by region and year')
plt.show()
```

```python
df6=df[['Year','Region','Count']]
df6=df6.groupby(['Year','Region']).sum().sort_values(by='Year',ascending=False).head(40)
df6
```

```python
df6.plot(kind='bar',figsize=(12,6),colormap='viridis')
plt.xlabel('Year Region')
plt.ylabel('Count')
plt.title('Sum of rate by region and year')
plt.show()

```

```python
df7=df.groupby('Subregion')['Count'].mean().sort_values(ascending=False).round(2)
df7
```

```python
df7.index
```

```python
df7.values
```

```python
d1={
    'Category':df7.index,
    'Value':df7.values,
    'Info':df7.values
}
df=pd.DataFrame(d1)
df
```

```python
fig=px.treemap(df,path=['Category'],values='Value',title='Treemap')
fig.update_traces(hovertemplate='Category:%{label}<br>Values:%{value}')
fig.show()
```
